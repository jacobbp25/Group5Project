```python
import zipfile
import os.path

directory = 'data'
files = [
    '{}/cardTransaction.csv.zip'.format(directory),
    '{}/cardTransaction2.csv.zip'.format(directory),
    '{}/VCardTransactions-NoName.csv.zip'.format(directory),
    '{}/VCardTransactions-NoName2.csv.zip'.format(directory)
]

for file in files:
    csvFile = file.replace('.zip','')
    
    if(not os.path.exists(csvFile)):
        print('File DOES NOT exist {} - un-zipping'.format(csvFile))
        with zipfile.ZipFile(file, 'r') as zip_ref:
            zip_ref.extractall(directory)
    else:
        print('File exists {} - NOT un-zipping'.format(csvFile))


transFile = '{}/trans.csv'.format(directory)
os.makedirs(os.path.dirname(transFile), exist_ok=True)

if(not os.path.exists(transFile)):
    print('File NOT exists {} - merging'.format(transFile))
    with open(files[0].replace('.zip','')) as fp:
        data = fp.read()

    with open(files[1].replace('.zip','')) as fp:
        data2 = fp.read()

    data += "\n"
    data += data2

    with open (transFile, 'w') as fp:
        fp.write(data)
else:
    print('File exists {} - NOT merging'.format(transFile))
        
viewFile = '{}/vtrans.csv'.format(directory)
os.makedirs(os.path.dirname(transFile), exist_ok=True)
if(not os.path.exists(viewFile)):
    print('File NOT exists {} - merging'.format(viewFile))
    with open(files[2].replace('.zip','')) as fp:
        data = fp.read()

    with open(files[3].replace('.zip','')) as fp:
        data2 = fp.read()

    data += "\n"
    data += data2

    with open (viewFile, 'w') as fp:
        fp.write(data)
else:
    print('File exists {} - NOT merging'.format(viewFile))
```

```python
import pandas as pd

trans = pd.read_csv(transFile)
vtrans = pd.read_csv(viewFile)
```
